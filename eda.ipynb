```python
plt.figure(figsize=(15,10))
sns.countplot(x=df.month_added, \
              order=["january", "february", "march", "april", "may", "june", \
                    "july", "august", "september", "october", "november", "december"]
             );
```

```python
df_us = pd.DataFrame(columns=df.columns)
```

```python
for i in range(len(df)):
    if (df.country_filled == "united states")[i]:
        df_us.loc[len(df_us)] = df.iloc[i]
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df_us.month_added, \
              order=["january", "february", "march", "april", "may", "june", \
                    "july", "august", "september", "october", "november", "december"]
             );
```

Although overall there seems to be no significant information in the month added, there may be trends within countries

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df.rating_filled);
```
